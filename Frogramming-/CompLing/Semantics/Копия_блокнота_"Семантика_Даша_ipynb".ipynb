````python
nltk.download('stopwords')
stops_ru = set(stopwords.words('russian'))
stops_en = set(stopwords.words('english'))
````

````python
def lemmas_sent(text, nlp_module, stops):
    words = nlp_module(text)
    all_sent = []
    for sent in words.sents:
        list_words = []
        for token in sent:
            lemm = token.lemma_
            if token.pos_ not in ['PUNCT', 'SPACE'] and lemm not in stops:
                list_words.append(lemm)
        list_words = ' '.join(list_words)
        all_sent.append(list_words)

    return '\n'.join(all_sent)
````

```
doc2bow(document, allow_update=False, return_missing=False)
Convert document into the bag-of-words (BoW) format = list of (token_id, token_count) tuples.

Parameters
document (list of str) – Input document.

allow_update (bool, optional) – Update self, by adding new tokens from document and updating internal corpus statistics.

return_missing (bool, optional) – Return missing tokens (tokens present in document but not in self) with frequencies?

Returns
list of (int, int) – BoW representation of document.

list of (int, int), dict of (str, int) – If return_missing is True, return BoW representation of document + dictionary with missing tokens and their frequencies.
```

````python
# теперь сделаем это системой, муаххаха
# в dict_ans уже лежат индексы всего, что нам нужно для 1 части
def get_count(books, list_words_parts):
    '''
    list_words_parts = split lemmatized text of each book
    in common format: [[1 book], [2 book]...]
    books = joined combined (?) lemmatized text of all books together
    in format:
    1 sent
    2 sent
    ...
    '''
    books = books.split()
    base_dct = Dictionary([books])
    parts_dct = [base_dct.doc2bow(list_words_parts[i])
                                        for i in range(len(list_words_parts))]
    # in our context length of list_words_parts = 10

    #dict_ans = {}
    #for i, j in enumerate(books.split()):  # работает, хоть и долго грузится
     #   key = base_dct.doc2idx(books.split())[i]
      #  dict_ans[j] = key

    ans = []
    for i in range(len(parts_dct)):
        ans.append({j[0]: j[1] for j in parts_dct[i]})

    l = pd.DataFrame({i: None for i in sorted(set(books))}, list(range(10)))
    sh = sorted(dict_ans.items(), key=lambda x: x[0])
    cringe = {i[1]: i[0] for i in sh}
    for_ans = []

    for i in ans:
        avg = {}
        for j in i:
            avg[cringe[j]] = i[j]
        for_ans.append(avg)

    return pd.DataFrame(for_ans)

ans = get_count(books, list_words_parts)
ans.update(ans.select_dtypes(include=[np.number]).fillna(0))
ans = ans.astype(int)
ans
````
